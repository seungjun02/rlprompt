```python
# prompt: /content/drive/MyDrive/rl-prompt-main/examples/few-shot-classification/outputs/2025-05-16/15-34-41/outputs/eval 경로에 있는 json 파일 전부다 읽어서 df로 저장하기

import glob
import json
import pandas as pd

# Specify the directory path
directory_path = '/content/drive/MyDrive/rl-prompt-main/examples/few-shot-classification/outputs/2025-05-18/12-47-25/outputs/eval'

# Get a list of all JSON files in the directory
json_files = glob.glob(f'{directory_path}/*.json')

# Initialize an empty list to store dataframes
dfs = []

# Read each JSON file and append it to the list of dataframes
for file_path in json_files:
    with open(file_path, 'r') as f:
        data = json.load(f)
        # Assuming each JSON file contains a list of records or a single record that can be a row
        # Adjust the following line if your JSON structure is different
        if isinstance(data, list):
            df = pd.DataFrame(data)
        else:
            df = pd.DataFrame([data]) # Treat a single dict as a single row
        dfs.append(df)

# Concatenate all dataframes into a single dataframe
if dfs:
    all_data_df = pd.concat(dfs, ignore_index=True)
    print(all_data_df.head()) # Print the head of the combined dataframe
else:
    print("No JSON files found in the specified directory.")
    all_data_df = pd.DataFrame() # Create an empty dataframe if no files were found
```

```python
# prompt: score값을 인덱스 [0]로 참조해서 정렬좀 해라

all_data_df['score_value'] = all_data_df['scores'].apply(lambda x: x[0])
sorted_df = all_data_df.sort_values(by='score_value', ascending=False)
sorted_df = sorted_df.drop(columns=['score_value'])
sorted_df
```
